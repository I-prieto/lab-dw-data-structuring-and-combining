2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Create a pivot table showing average CLV per gender and education level
pivot_table = df_file4.pivot_table(
    index='gender',
    columns='education',
    values='customer_lifetime_value',
    aggfunc='mean'
)

# Round for readability
pivot_table = pivot_table.round(2)

# Display the result
print("Average Customer Lifetime Value by Gender and Education Level:")
print(pivot_table)
```

```python
#Insight 2:
#The pivot table shows that female customers generally have a higher average customer lifetime value than male customers across most education levels. The highest value is observed among female customers with a high school education or below, suggesting that this group may be particularly loyal or engaged. Interestingly, female customers with doctorate degrees have the lowest average lifetime value, which could indicate different purchasing behaviours or shorter engagement. Among male customers, those with college and master’s degrees show relatively strong values, but none exceed the top-performing female segments. Overall, education level does not consistently correlate with higher lifetime value, and gender differences are notable, especially at the extremes. Recommendation consider targeted marketing strategies that leverage these insights to enhance customer retention and value.
```
